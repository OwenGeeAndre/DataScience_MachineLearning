# Feature Selection - Standardization

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
````

## Load the Data

````python
df = pd.read_csv(r'/content/1.02.+Multiple+linear+regression.csv')
df.head(2)
````

# Multiple Linear Regression

## Independent and Dependent Variables


````python
target = "GPA"
x = df.drop(columns=target)
y = df[target]
````

````python
x.head(3)
````

## Standardization

````python
from sklearn.preprocessing import StandardScaler
````

````python
scaler = StandardScaler()
````

````python
scaler.fit(x)
````

````python
x_scaled = scaler.transform(x)
````

# Regression with Scaled Features

````python
reg = LinearRegression()
reg.fit(x_scaled, y)
````

````python
reg.coef_
````

````python
reg.intercept_
````

# Creating a Summary Table

````python
reg_summary = pd.DataFrame([["Bias"],["SAT"],["Rand 1,2,3"]], columns =["Features"])
reg_summary["Weights"] = reg.intercept_, reg.coef_[0], reg.coef_[1]
````

````python
reg_summary
````



```
For a scaled data: The closer a weight is to O, the smaller its impact;
The bigger the weight, the bigger its impact
```



# Making Predictions (with standardized weights)

````python
new_data = pd.DataFrame(data=[[2100,2],[1500,1]], columns=["SAT", "Rand 1,2,3"])
new_data
````

````python
reg.predict(new_data)
````

> The input data has to be in the same form as the data used to train the model

````python
new_data_scaled = scaler.transform(new_data)
new_data_scaled
````

````python
reg.predict(new_data_scaled)
````

# Modelling (without `Rand 1,2,3`)

````python
new_reg = LinearRegression()
x_simple = x_scaled[:,0].reshape(-1,1)
new_reg.fit(x_simple, y)
````

````python
new_reg.predict(new_data_scaled[:,0].reshape(-1,1))
````
